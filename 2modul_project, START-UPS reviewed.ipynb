- –û—Ü–µ–Ω–∏—Ç–µ, –Ω–∞—Å–∫–æ–ª—å–∫–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –ø–æ–ª–Ω–∞. –ò—Å–ø–æ–ª—å–∑—É—è –¥–∞—Ç–∞—Å–µ—Ç—ã `people.csv` –∏ `education.csv`, —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –≤—Å–µ –∫–æ–º–ø–∞–Ω–∏–∏ –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –≥—Ä—É–ø–ø –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –∏ –æ—Ü–µ–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω—é—é –¥–æ–ª—é —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –±–µ–∑ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ –≤ –∫–∞–∂–¥–æ–π –∏–∑ –≥—Ä—É–ø–ø. –û–±–æ—Å–Ω—É–π—Ç–µ –≤—ã–±—Ä–∞–Ω–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã –≥—Ä—É–ø–ø.
- –û—Ü–µ–Ω–∏—Ç–µ, –≤–æ–∑–º–æ–∂–Ω–æ –ª–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏—è –ø—Ä–∏—Å–æ–µ–¥–∏–Ω–∏—Ç—å –∫ —ç—Ç–∏–º —Ç–∞–±–ª–∏—Ü–∞–º –µ—â—ë –∏ —Ç–∞–±–ª–∏—Ü—É `degrees.csv`.

–û–±—ä–µ–¥–∏–Ω–∏–º –¥–∞—Ç–∞—Å–µ—Ç—ã people –∏ education –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º: id=person_id.

–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü is_educated - –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å, —á—Ç–æ —É —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞ –æ—Ç–º–µ—á–µ–Ω–∞ –¥–∞—Ç–∞ –æ–∫–æ–Ω—á–∞–Ω–∏—è –æ–±—Ä–∞–∑–æ–≤–∞—Ç–µ–ª—å–Ω–æ–≥–æ —É—á—Ä–µ–∂–¥–µ–Ω–∏—è –∏–ª–∏ —Ç–∏–ø –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è.

```python
total_education_clean['is_educated']=0
total_education_clean.loc[total_education_clean['graduated_at'].notna() | total_education_clean['degree_type'].notna(), 'is_educated']=1
total_education_clean.head()
```

–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –∫–æ–º–ø–∞–Ω–∏—è–º –∏ —É–∑–Ω–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Å–µ—Ö —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ —Å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–µ–º –≤ –∫–∞–∂–¥–æ–π –∫–æ–º–ø–∞–Ω–∏–∏.

__–ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –∫–æ–º–ø–∞–Ω–∏–π –ø–æ —á–∏—Å–ª–µ–Ω–Ω–æ—Å—Ç–∏ –ø–µ—Ä—Å–æ–Ω–∞–ª–∞:__

–ò–ü/—Å—Ç–∞—Ä—Ç–∞–ø: 1 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫ (–æ—Ç–¥–µ–ª—å–Ω–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è –¥–ª—è –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–ø—Ä–∏–Ω–∏–º–∞—Ç–µ–ª–µ–π –∏ –Ω–∞—á–∏–Ω–∞—é—â–∏—Ö —Å—Ç–∞—Ä—Ç–∞–ø–æ–≤)

–ú–∏–∫—Ä–æ–ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏—è: –æ—Ç 2 –¥–æ 15 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤

–ú–∞–ª—ã–µ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏—è: –æ—Ç 16 –¥–æ 100 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤

–°—Ä–µ–¥–Ω–∏–µ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏—è: –æ—Ç 101 –¥–æ 250 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤

–ö—Ä—É–ø–Ω—ã–µ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏—è (—Å–≤—ã—à–µ 250 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤) –≤ –≤—ã–±–æ—Ä–∫–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç, –ø–æ—ç—Ç–æ–º—É –≤ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ —É—á–∏—Ç—ã–≤–∞—é—Ç—Å—è.

–î–ª—è –∞–Ω–∞–ª–∏–∑–∞ –±—É–¥–µ—Ç —Å–æ–∑–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è, –¥–æ–±–∞–≤–ª—è—é—â–∞—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π —Å—Ç–æ–ª–±–µ—Ü –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º. –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –±—É–¥–µ—Ç —Ä–∞—Å—Å—á–∏—Ç–∞–Ω –ø—Ä–æ—Ü–µ–Ω—Ç —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ —Å –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–Ω—ã–º –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º –≤ –∫–∞–∂–¥–æ–π —Ä–∞–∑–º–µ—Ä–Ω–æ–π –≥—Ä—É–ø–ø–µ, —á—Ç–æ –ø–æ–∑–≤–æ–ª–∏—Ç –≤—ã—è–≤–∏—Ç—å –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É —Ä–∞–∑–º–µ—Ä–æ–º –∫–æ–º–ø–∞–Ω–∏–∏ –∏ —É—Ä–æ–≤–Ω–µ–º –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤.

```python
def categories(x):
  if x['count']>100:
    return '–°—Ä–µ–¥–Ω–µ–µ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–µ'
  elif 100>=x['count']>15:
    return '–ú–∞–ª–æ–µ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–µ'
  elif 15>=x['count']>1:
    return '–ú–∏–∫—Ä–æ–ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–µ'
  else:
    return '–ò–ü/—Å—Ç–∞—Ä—Ç–∞–ø'    

group_df['company_category']=group_df.apply(categories, axis=1)

group_df['perc']=round(group_df['sum']/group_df['count']*100,2)
display(group_df)
```

–°–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, –æ–ø—Ä–µ–¥–µ–ª–∏–º –¥–æ–ª—é —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ —Å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º, —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–≤ –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –∫–æ–º–ø–∞–Ω–∏–π. 

```python
company_category_df = group_df.pivot_table(values=['sum', 'count'], index='company_category', aggfunc='sum').reset_index()

company_category_df['percentage'] = round(company_category_df['sum'] / company_category_df['count'] * 100,2)

company_category_df = company_category_df.sort_values(by='percentage', ascending=False)

display(company_category_df)
```
