–∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—É –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞ –∑–∞–≤–µ–¥–µ–Ω–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏: –µ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è,

–∫–∞—Ç–µ–≥–æ—Ä–∏—è –∏ –ø—Ä–∏–∑–Ω–∞–∫, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∑–∞–≤–µ–¥–µ–Ω–∏—è –∫—Ä—É–≥–ª–æ—Å—É—Ç–æ—á–Ω—ã–º. –í—ã–±–µ—Ä–∏—Ç–µ —Å–∞–º—É—é —Å–∏–ª—å–Ω—É—é —Å–≤—è–∑—å –∏
–ø—Ä–æ–≤–µ—Ä—å—Ç–µ –µ—ë.

```python
df_rating = df[['category','district','chain','seats','price','is_24_7','rating']]
df_rating_corr = df_rating.phik_matrix()
rating_heatmap = df_rating_corr
sns.heatmap(
 rating_heatmap,
 annot=True,
 fmt='.2f',
 cmap='coolwarm',
 linewidths=1,
 cbar=True
)
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ phi_k –¥–ª—è –¥–∞–Ω–Ω—ã—Ö rating')
print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–º phi_k –¥–ª—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π rating')
df_rating_corr.loc[df_rating_corr.index != 'rating'][['rating']].sort_values(by='rating',ascending=False)
```

### –í–æ–ø—Ä–æ—Å ‚Ññ7
–°–≥—Ä—É–ø–ø–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—è–º –∑–∞–≤–µ–¥–µ–Ω–∏–π –∏ –Ω–∞–π–¥–∏—Ç–µ —Ç–æ–ø-15 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö —Å–µ—Ç–µ–π –≤ –ú–æ—Å–∫–≤–µ. –î–ª—è –Ω–∏—Ö
–ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞. –ü–æ–¥ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å—é –ø–æ–Ω–∏–º–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π —ç—Ç–æ–π
—Å–µ—Ç–∏ –≤ —Ä–µ–≥–∏–æ–Ω–µ. –ö –∫–∞–∫–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∑–∞–≤–µ–¥–µ–Ω–∏–π –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è? –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ–ø—Ä–æ–≤–æ–¥–∏—Ç–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–º–∏
–≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è–º–∏.

```python
df_chain = df[df['chain'] == 1]

chain_count_rating = df_chain.groupby(['name','category']).agg(
    name_count=('name', 'count'),
    avg_rating=('rating', 'mean')
).sort_values('name_count', ascending=False).reset_index()

top_15 = round(chain_count_rating.head(15),2)

sns.barplot(
    data=top_15,
    x='name',
    y='name_count',
    hue='category'
)

plt.title('–¢–æ–ø-15 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π —Å–µ—Ç–µ–π –≤ –ú–æ—Å–∫–≤–µ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ)')
plt.legend(title='–ö–∞—Ç–µ–≥–æ—Ä–∏—è –∑–∞–≤–µ–¥–µ–Ω–∏—è')
plt.xlabel('–ù–∞–∑–≤–∞–Ω–∏–µ —Å–µ—Ç–∏')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π')
plt.xticks(rotation=90)
plt.grid(axis='y')

plt.show()

display('–¢–æ–ø-15 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π —Å–µ—Ç–µ–π –≤ –ú–æ—Å–∫–≤–µ')
top_15
```

```python
top_15_sorted = top_15.sort_values('avg_rating', ascending=False)

display('–¢–æ–ø-15 —Å–µ—Ç–µ–π –ú–æ—Å–∫–≤—ã, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ —Ä–µ–π—Ç–∏–Ω–≥—É')
display(top_15_sorted)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_15_sorted,
    x='name',
    y='avg_rating',
    hue='category',
    order=top_15_sorted['name']  
)

plt.title('–¢–æ–ø-15 —Å–µ—Ç–µ–π –ú–æ—Å–∫–≤—ã –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É —Ä–µ–π—Ç–∏–Ω–≥—É', fontsize=14)
plt.xlabel('–ù–∞–∑–≤–∞–Ω–∏–µ —Å–µ—Ç–∏', fontsize=12)
plt.ylabel('–°—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.legend(title='–ö–∞—Ç–µ–≥–æ—Ä–∏—è', bbox_to_anchor=(1.05, 1))
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()

for index, value in enumerate(top_15_sorted['avg_rating']):
    plt.text(index, value + 0.02, f'{value:.2f}', ha='center', va='bottom', fontsize=9)

plt.show()
```
